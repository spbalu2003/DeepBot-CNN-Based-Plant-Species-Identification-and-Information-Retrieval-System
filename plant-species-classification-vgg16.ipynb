<h3 id="importing-libraries" style="font-family: 'Georgia', serif; font-size: 24px; color: #333;">4.1 Importing Libraries</h3>

```python
# Importing necessary libraries
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.utils import class_weight
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.applications import VGG16
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
import tensorflow as tf
```

<h3 id="loading-and-preprocessing-the-data" style="font-family: 'Georgia', serif; font-size: 24px; color: #333;">4.2 Loading and Preprocessing the Data</h3>

```python
# Setting paths
root_dir = '/kaggle/input/house-plant-species/house_plant_species'
categories = os.listdir(root_dir)

# List all image paths and labels
image_paths = []
labels = []

for category in categories:
    for image_name in os.listdir(os.path.join(root_dir, category)):
        image_paths.append(os.path.join(root_dir, category, image_name))
        labels.append(category)

# Encode labels
label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(labels)

# Image preprocessing function
IMG_SIZE = 128

def preprocess_image(image_path):
    img = cv2.imread(image_path)
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
    img = img / 255.0  # Normalize the image
    return img

# Apply preprocessing
train_images = np.array([preprocess_image(path) for path in image_paths])
```

<h3 id="train-test-split" style="font-family: 'Georgia', serif; font-size: 24px; color: #333;">4.3 Train-Test Split</h3>

```python
# Train-test split
train_paths, test_paths, train_labels, test_labels = train_test_split(
    image_paths, encoded_labels, test_size=0.2, random_state=42)

train_images = np.array([preprocess_image(path) for path in train_paths])
test_images = np.array([preprocess_image(path) for path in test_paths])
```

<h3 id="data-augmentation" style="font-family: 'Georgia', serif; font-size: 24px; color: #333;">4.4 Data Augmentation</h3>

```python
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Fit the data augmentation generator
datagen.fit(train_images)
```

<h3 id="handling-class-imbalance" style="font-family: 'Georgia', serif; font-size: 24px; color: #333;">4.5 Handling Class Imbalance</h3>

```python
class_weights = dict(enumerate(class_weight.compute_class_weight(
    class_weight='balanced', 
    classes=np.unique(train_labels), 
    y=train_labels)))
```

<h3 id="building-the-cnn-model" style="font-family: 'Georgia', serif; font-size: 24px; color: #333;">4.6 Building the CNN Model</h3>

```python

# Load pre-trained VGG16 model
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(IMG_SIZE, IMG_SIZE, 3))
base_model.trainable = True  

# Freeze all layers except for the top 4 layers
for layer in base_model.layers[:-4]:
    layer.trainable = False

# Define CNN model with VGG16 as base
model = Sequential([
    base_model,
    Flatten(),
    Dense(512, activation='relu'), 
    BatchNormalization(),           
    Dropout(0.5),
    Dense(256, activation='relu'),
    BatchNormalization(),
    Dropout(0.5),
    Dense(128, activation='relu'),
    BatchNormalization(),
    Dropout(0.5),
    Dense(47, activation='softmax') 
])
```

<h3 id="compiling-the-model" style="font-family: 'Georgia', serif; font-size: 24px; color: #333;">4.7 Compiling the Model</h3>

```python
# Compile the model with a lower learning rate
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

<h3 id="implementing-callbacks" style="font-family: 'Georgia', serif; font-size: 24px; color: #333;">4.8 Implementing Callbacks</h3>

```python
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Save the best model during training
model_checkpoint = ModelCheckpoint('best_model.keras', save_best_only=True, monitor='val_loss')

# Reduce learning rate when validation performance plateaus
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=1e-6)
```

<h3 id="training-the-model" style="font-family: 'Georgia', serif; font-size: 24px; color: #333;">4.9 Training the Model</h3>

```python
history = model.fit(datagen.flow(train_images, train_labels, batch_size=32),
                    validation_data=(test_images, test_labels),
                    epochs=50,
                    class_weight=class_weights,
                    callbacks=[early_stopping, model_checkpoint, reduce_lr])
```

<h3 id="evaluating-the-model" style="font-family: 'Georgia', serif; font-size: 24px; color: #333;">4.10 Evaluating the Model</h3>

```python
# Evaluate the model
model.evaluate(test_images, test_labels)

# Plot training history
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()
plt.show()

plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```

<h3 id="confusion-matrix-and-classification-report" style="font-family: 'Georgia', serif; font-size: 24px; color: #333;">4.11 Confusion Matrix and Classification Report</h3>

```python
# Confusion Matrix and Classification Report
pred_labels = np.argmax(model.predict(test_images), axis=1)

# Confusion Matrix (Normalized)
cm = confusion_matrix(test_labels, pred_labels, normalize='true')
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True, fmt='.2f', cmap='Blues')
plt.title('Normalized Confusion Matrix')
plt.show()

# Classification Report
print(classification_report(test_labels, pred_labels, target_names=label_encoder.classes_))
```

<h3 id="displaying-sample-predictions" style="font-family: 'Georgia', serif; font-size: 24px; color: #333;">4.12 Displaying Sample Predictions</h3>

```python
for i in range(10):
    plt.imshow(test_images[i])
    plt.title(f"Actual: {label_encoder.inverse_transform([test_labels[i]])[0]} | Predicted: {label_encoder.inverse_transform([pred_labels[i]])[0]}")
    plt.show()
```
